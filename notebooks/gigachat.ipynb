```python
from dotenv import load_dotenv
import os

load_dotenv()
API_KEY = os.getenv("API_KEY")
```

Проверка доступности API

```python
from langchain_gigachat.chat_models import GigaChat

giga = GigaChat(
    credentials=API_KEY,
    verify_ssl_certs=False,
)

print(giga.invoke("Hello, world!").content)
```

Список доступных моделей

```python
models = giga.get_models()

print("Доступные модели:\n")

for i, m in enumerate(models.data, 1):
    print(f"{i:2}. {m.id_:25} | owner: {m.owned_by}")
```

Простой запрос на проверку работы LLM:

```python
print(giga.invoke("Сколько будет корень из 2 округленно до 2 знаков ?").content)
```
