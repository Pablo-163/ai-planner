# Demo AI-планировщика здач

Planning-агент на базе LangGraph и GigaChat.

Что умеет:
- извлекать параметры из цели пользователя
- проверять, хватает ли данных
- декомпозировать цель на подзадачи
- строить итоговый план

```python
# call root files
import sys
from pathlib import Path

sys.path.append(str(Path().resolve().parent))

from graph import graph
```

```python
def make_initial_state(user_goal: str) -> dict:
    return {
        "user_goal": user_goal,
        "deadline": None,
        "available_time_per_day": None,
        "user_level": None,
        "missing_fields": [],
        "subtasks": [],
        "draft_plan": "",
        "final_plan": "",
    }
```

## Базовый сценарий

```python
state = make_initial_state(
    "Хочу подготовиться к Python-собеседованию за 2 недели, "
    "могу заниматься 4 часа в день, уровень junior"
)

result = graph.invoke(state)
print(result["final_plan"])
```

## Данных мало

```python
state = make_initial_state("Хочу выучить SQL")

result = graph.invoke(state)
print(result["final_plan"])
```

## Промежуточные результаты

```python
state = make_initial_state(
    "Хочу подготовиться к Python-собеседованию за 2 недели, "
    "могу заниматься 4 часа в день, уровень junior"
)

result = graph.invoke(state)

print("deadline:", result["deadline"])
print("available_time_per_day:", result["available_time_per_day"])
print("user_level:", result["user_level"])
print("missing_fields:", result["missing_fields"])

print("\nsubtasks:")
for i, task in enumerate(result["subtasks"], start=1):
    print(f"{i}. {task}")

print("\nfinal_plan:\n")
print(result["final_plan"])
```
